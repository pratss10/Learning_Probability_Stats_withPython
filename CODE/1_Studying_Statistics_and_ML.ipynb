```python
ad = pd.read_csv('Advertising.csv', index_col = 0)
```

```python
ad.head(10)
```

```python
ad.describe()
```

```python
ad['stv'] = ad['sales']*ad['TV']
ad['sqtv'] = ad['TV']**2
n = ad.count()
```

```python
mtv = (n * ad['stv'].sum() - (ad['sales'].sum() * ad['TV'].sum())) / (n* ad['sqtv'].sum() - (ad['TV'].sum())**2 )
mtv
```

```python
n = len(ad)  # number of observations
mean_tv = ad['TV'].mean()  # mean of TV
mean_sales = ad['sales'].mean()  # mean of sales

numerator = n * ((ad['TV'] - mean_tv) * (ad['sales'] - mean_sales)).sum()

denominator = n * ((ad['TV'] - mean_tv) ** 2).sum()

slope = numerator / denominator

print(slope)
```

```python
intercept = ad['sales'].mean() - slope*ad['TV'].mean()
intercept
```

```python
def pred_tv_sales(tv):
  return intercept + slope*tv
```

```python
print( pred_tv_sales(1000))
```
