```python
import numpy as np
import pandas as pd
from matplotlib.pyplot import subplots
import statsmodels.api as sm
from ISLP import load_data
from ISLP.models import (ModelSpec as MS, summarize)

from ISLP import confusion_table
from ISLP.models import contrast
from sklearn.discriminant_analysis import (LinearDiscriminantAnalysis as LDA, QuadraticDiscriminantAnalysis as QDA)
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df = load_data('Caravan')
df
```

```python
df.columns #purchase Caravan insurance policy
```

```python
y = df.Purchase
y.value_counts()
```

```python
348/5474
```

```python
# ~6% people buy
```

```python
x = df.drop(columns = ['Purchase'])
```

```python
#since knn only cares about the distance it will prioritze large numbers without any context
#so we will standardize all of the predictor inputs using standard scaler class
```

```python
scaler = StandardScaler(with_mean = True,
                        with_std = True,
                        copy = True)
```

```python
scaler.fit(x)
X = scaler.transform(x)
#mean = 0, std = 1
```

```python
df.shape
```

```python
(X_train,
X_test ,
y_train ,
y_test) = train_test_split(X,
y , test_size=822, random_state=0)
```

```python
knn1 = KNeighborsClassifier(n_neighbors = 1)
knn1_pred = knn1.fit(X_train, y_train).predict(X_test)
np.mean(y_test != knn1_pred) , np.mean(y_test != "No")
```

```python
#11.67 is bad, because i couldve gotten 7% by just answering NO each time
#However for an insurance company,
#it interests lie in people who are more likely to buy it and hence our predictor should be correct on people tat will buy
```

```python
confusion_table(knn1_pred, y_test)
```

```python
9/(9+46)
```

```python
#16% yes is good compared to 6% random gussing
```

Tuning KNN by finding optimum K

```python
for K in range(1,6):
  knn = KNeighborsClassifier(n_neighbors = K)
  knn_pred = knn.fit(X_train, y_train).predict(X_test)
  c = confusion_table(knn_pred, y_test)
  pred = c.loc['Yes'].sum()
  actual = c.loc['Yes','Yes']
  print( K, actual, pred, actual*100/pred)
```

```python
# K=2 Seems to be the best
```

Comparing our Results with Logistic Regression

```python
logit = LogisticRegression(C=1e10, solver='liblinear')
logit.fit(X_train, y_train)
logit_pred = logit.predict_proba(X_test)
logit_labels = np.where(logit_pred[:,1] > .5, 'Yes', 'No')
confusion_table(logit_labels, y_test)
```

```python
#LOL, it will get 6% correct everytime but we care about poeple that want to buy and not the ones that are gonna say no anyways
#so trying with p > 0.25 as yes
```
